```python
# Bivariate Analysis
# Correlation Analysis
import numpy as np
numeric_columns = df_high.select_dtypes(include=[np.number]).columns
correlation_matrix = df_high[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Contingency Tables
contingency_table = pd.crosstab(df_high['parent_type'], df_high['commodity'])
print(contingency_table)
```

```python
# PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_high[numeric_columns])

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
df_high['pca_one'] = pca_result[:,0]
df_high['pca_two'] = pca_result[:,1]

sns.scatterplot(x='pca_one', y='pca_two', data=df_high)
plt.show()

# Regression Analysis
import statsmodels.api as sm

X = df_high[['production_value', 'flaring_emissions_MtCO2', 'venting_emissions_MtCO2']]
Y = df_high['total_emissions_MtCO2e']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
print(model.summary())

# Clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
df_high['cluster'] = kmeans.fit_predict(scaled_data)

sns.scatterplot(x='pca_one', y='pca_two', hue='cluster', data=df_high)
plt.show()
```
